```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
df.head()
```

```python
#plt.figure(figsize=(5, 8))
sns.rugplot(x='salary', data=df, height=0.5)
```

```python
sns.displot(x='salary',data=df)
```

```python
sns.set(style='darkgrid')

sns.displot(x='salary',data=df, bins=20, color='red', edgecolor='blue', kde=True, rug=True)
```

```python
sns.histplot(data=df, x='salary')
```

```python
sns.kdeplot(data=df, x='salary')
```

```python
np.random.seed(42)
sample_ages = np.random.randint(0, 100, 200)
```

```python
sns.kdeplot(data=df, x='salary')
```

```python
sample_ages = pd.DataFrame(sample_ages, columns=['age'])
```

```python
sns.rugplot(data=sample_ages, x='age')
```

```python
sns.displot(data=sample_ages, x='age', rug=True, bins=30, kde=True)
```

```python
sns.kdeplot(data=sample_ages, x='age', clip=[0, 100])
```

```python
sns.kdeplot(data=sample_ages, x='age', clip=[0, 100], bw_adjust=0.6, shade=True)
```
