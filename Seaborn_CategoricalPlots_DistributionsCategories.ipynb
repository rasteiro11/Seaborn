```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.boxplot(data=df, y='math score', x = 'test preparation course')
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.boxplot(data=df, y='reading score', x = 'test preparation course')
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.boxplot(data=df, y='reading score', x = 'parental level of education', hue='test preparation course', palette='Set2')
plt.legend(bbox_to_anchor=(1.05, 0.5))
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.violinplot(data=df, x='reading score', y= 'parental level of education', hue='test preparation course')
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.violinplot(data=df, x='reading score', y= 'parental level of education', hue='test preparation course', split=True)
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.violinplot(data=df, x='reading score', y= 'parental level of education', hue='test preparation course', split=True, inner='quartile')
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.violinplot(data=df, x='reading score', y= 'parental level of education', inner='stick')
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.violinplot(data=df, x='reading score', y= 'parental level of education', bw=0.1)
```

```python
plt.figure(figsize=(8,4), dpi=200)
sns.swarmplot(data=df, x='math score', size=2)
```

```python
plt.figure(figsize=(8,4), dpi=200)
sns.swarmplot(data=df, x='math score', y='gender', size=2, hue='test preparation course', dodge=True)
plt.legend(bbox_to_anchor=(1.35, 0.5))
```

```python
sns.boxenplot(x='math score', y='test preparation course', data=df, hue='gender')
```
